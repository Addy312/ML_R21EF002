```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('cars.csv')
```

```python
# Plot 1: Pair plot for all features
sns.pairplot(data, hue='year')
plt.title('Pair Plot for cars Dataset')
plt.show()
```

```python
# Plot 2: Box plot for Selling price by km driven
sns.boxplot(x='selling_price', y='km_driven', data=data)
plt.title('Box Plot for Selling price by km driven')
plt.show()
```

```python
# Plot 3: Violin plot for Selling price by km driven
sns.violinplot(x='selling_price', y='km_driven', data=data)
plt.title('Violin Plot for Selling price by km driven')
plt.show()
```

```python
# Plot 4: Histogram of Sepal Length
plt.hist(data[data['transmission'] == 'Manual']['selling_price'], alpha=0.5, label='Manual')
plt.hist(data[data['transmission'] == 'Automatic']['selling_price'], alpha=0.5, label='Auto')
plt.title('Histogram of selling_price by transmission')
plt.legend()
plt.show()
```

```python
# Plot 5: Scatter plot for selling_price vs km_driven
plt.scatter(data[data['transmission'] == 'Manual']['selling_price'], data[data['transmission'] == 'Manual']['km_driven'], label='Man')
plt.scatter(data[data['transmission'] == 'Automatic']['selling_price'], data[data['transmission'] == 'Automatic']['km_driven'], label='Auto')
plt.xlabel('selling_price')
plt.ylabel('km_driven')
plt.title('Scatter Plot of selling_price vs km_driven by transmission')
plt.legend()
plt.show()

```

```python
# Plot 6: Bar plot for Species counts
transmission_counts = data['transmission'].value_counts()
transmission_counts.plot(kind='bar', color='skyblue')
plt.title('Bar Plot of transmission Counts')
plt.xlabel('transmission')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 8: Petal selling price vs km driven
sns.scatterplot(x='selling_price', y='km_driven', hue='transmission', data=data)
plt.title('Scatter Plot of selling price vs km driven by transmission')
plt.show()
```

```python
# Plot 9: Petal Length distribution by Species
sns.kdeplot(data[data['transmission'] == 'Manual']['selling_price'], label='Manual')
sns.kdeplot(data[data['transmission'] == 'Automatic']['selling_price'], label='Auto')
plt.title('selling_price Distribution by transmission')
plt.legend()
plt.show()
```

```python

#plot 10
sns.swarmplot(x='transmission', y='selling_price', data=data, palette="Set2")
plt.title('Bee Swarm Plot for selling_price by transmission')
plt.xlabel('transmission')
plt.ylabel('selling_price')
plt.show()
```
